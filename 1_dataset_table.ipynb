```python
# Define the keyword to search in title or description
SEARCH_TERM = "*"  # Use "*" to get all videos from the institution

# Define the institution to filter by
INSTITUTION = "Netherlands Institute for Sound & Vision"

# API endpoint
API_URL = "https://api.europeana.eu/record/v2/search.json"

# Number of videos per page and total pages
RESULTS_PER_PAGE = 1000
TOTAL_VIDEOS = 100000  # Limit total results
TOTAL_PAGES = TOTAL_VIDEOS // RESULTS_PER_PAGE  # Number of pages needed

# Initialize cursor and results list
cursor = "*"
all_results = []

# Loop through pages
for page in range(TOTAL_PAGES):
    print(f"🔄 Fetching page {page + 1}...")

    # API parameters for cursor-based pagination
    params = {
        "wskey": API_KEY,  # Your API key
        "query": f"DATA_PROVIDER:(\"{INSTITUTION}\") AND {SEARCH_TERM}",
        "qf": ["TYPE:VIDEO"],  # Filter for video content
        "media": "true",  # Ensure media files are included
        "rows": RESULTS_PER_PAGE,  # Number of results per page
        "reusability": "open",  # Get only openly licensed content
        "cursor": cursor  # Cursor for pagination
    }

    # Send the GET request
    response = requests.get(API_URL, params=params)

    # Check response status
    if response.status_code == 200:
        data = response.json()

        # Extract video details
        for item in data.get("items", []):
            # 🔹 Convert all fields to **strings** safely
            title = item.get("title", "No Title")
            title = title[0] if isinstance(title, list) else str(title)  # Handle list case

            description = item.get("dcDescription", ["No Description"])
            description = description[0] if isinstance(description, list) and description else str(description)

            video_link = item.get("edmIsShownAt", ["No Video URL"])
            video_link = video_link[0] if isinstance(video_link, list) and video_link else str(video_link)

            provider = item.get("dataProvider", "Unknown Provider")
            provider = ", ".join(provider) if isinstance(provider, list) else str(provider)

            provider_info = item.get("provider", [])
            if isinstance(provider_info, list) and provider_info:
                alternative_provider = provider_info[0] if isinstance(provider_info[0], str) else provider_info[0].get("name", "Unknown")
            elif isinstance(provider_info, dict):
                alternative_provider = provider_info.get("name", "Unknown")
            else:
                alternative_provider = "Unknown"
            alternative_provider = ", ".join(alternative_provider) if isinstance(alternative_provider, list) else str(alternative_provider)

            # 🔹 Extract Europeana URL (guid or id)
            europeana_url = item.get("guid", item.get("id", "No Europeana URL"))

            # 🔹 Manually filter by institution name
            if INSTITUTION.lower() in provider.lower() or INSTITUTION.lower() in alternative_provider.lower():
                all_results.append({
                    "Title": title.strip(),
                    "Description": description.strip(),
                    "Video URL": video_link.strip(),
                    "Europeana URL": europeana_url.strip(),  # ✅ New Column
                    "Provider": provider.strip(),
                    "Alternative Provider": alternative_provider.strip()
                })

        # Get next cursor for pagination
        cursor = data.get("nextCursor")
        if not cursor:
            print("✅ No more pages available.")
            break  # Stop if there are no more results

    else:
        print(f"❌ Error fetching page {page + 1}: {response.status_code}")
        break  # Stop in case of API failure

# Convert results to DataFrame
df_results = pd.DataFrame(all_results)

# ✅ Drop empty columns if they exist
df_results = df_results.dropna(axis=1, how='all')

# Save to CSV
df_results.to_csv("video_search_results_paginated_clean.csv", index=False)

print(f"✅ Saved {len(df_results)} results to 'video_search_results_paginated_clean.csv'")
```

```python
import os
import requests
import pandas as pd
from urllib.parse import urlparse

# Europeana API key
VIDEO_FOLDER = "../videos"
DATA_FOLDER = "../data_info"
os.makedirs(VIDEO_FOLDER, exist_ok=True)
os.makedirs(DATA_FOLDER, exist_ok=True)

# Function to extract item ID from URL
def extract_item_id(europeana_url):
    path_parts = urlparse(europeana_url).path.strip("/").split("/")
    return "/".join(path_parts[-2:]) if len(path_parts) > 1 else None

# Function to download video from Europeana
def download_video(europeana_url):
    item_id = extract_item_id(europeana_url)
    if not item_id:
        print("Invalid URL format.")
        return
    
    metadata_url = f"https://api.europeana.eu/record/v2/{item_id}.json"
    response = requests.get(metadata_url, params={"wskey": API_KEY})
    
    if response.status_code == 200:
        data = response.json()
        aggregation = data.get("object", {}).get("aggregations", [{}])[0]
        video_url = aggregation.get("edmIsShownBy", "Unknown")
        
        if video_url != "Unknown":
            print(f"Downloading: {video_url}")
            video_response = requests.get(video_url, stream=True)
            if video_response.status_code == 200:
                video_filename = os.path.join(VIDEO_FOLDER, f"{item_id.replace('/', '_')}.mp4")
                with open(video_filename, "wb") as video_file:
                    for chunk in video_response.iter_content(chunk_size=1024):
                        video_file.write(chunk)
                print(f"Downloaded: {video_filename}")
            else:
                print(f"Failed to download video, Status Code: {video_response.status_code}")
        else:
            print("No video URL found.")
    else:
        print(f"Failed to fetch item {item_id} from Europeana API, Status Code: {response.status_code}")

# Example usage
europeana_url = "https://www.europeana.eu/en/item/2051906/data_euscreenXL_https___www_openbeelden_nl_media_652616"
download_video(europeana_url)
```

```python
import requests
import pandas as pd

# Define the keyword to search in title or description
SEARCH_TERM = "*"  # Change this to your desired keyword

# Define the institution to filter by
INSTITUTION = "Netherlands Institute for Sound & Vision"

# API endpoint
API_URL = "https://api.europeana.eu/record/v2/search.json"

# Parameters for the API request
params = {
    "wskey": API_KEY,  # Your API key
    "query": SEARCH_TERM,  # Search term in title or description
    "qf": ["TYPE:VIDEO"],  # Filter for video content
    "media": "true",  # Ensure media files are included
    "rows": 100,  # Increase number of results (adjust as needed)
    "reusability": "open"  # Get only openly licensed content
}

# Send the GET request
response = requests.get(API_URL, params=params)

# Check response status
if response.status_code == 200:
    data = response.json()

    # Extract video details
    results = []
    for item in data.get("items", []):
        title = item.get("title", "No Title")

        # Safely extract description (handle missing fields)
        description_list = item.get("dcDescription", ["No Description"])
        description = description_list[0] if isinstance(description_list, list) and description_list else "No Description"

        # Safely extract video link
        video_link = item.get("edmIsShownAt", "No Video URL")

        # 🔹 Check for institution name in different fields and ensure they are strings
        provider = item.get("dataProvider", "Unknown Provider")
        if isinstance(provider, list):  # If it's a list, join into a string
            provider = ", ".join(provider)

        provider_info = item.get("provider", [])
        if isinstance(provider_info, list) and provider_info:
            alternative_provider = provider_info[0] if isinstance(provider_info[0], str) else provider_info[0].get("name", "Unknown")
        elif isinstance(provider_info, dict):
            alternative_provider = provider_info.get("name", "Unknown")
        else:
            alternative_provider = "Unknown"

        if isinstance(alternative_provider, list):
            alternative_provider = ", ".join(alternative_provider)

        source = item.get("source", "Unknown Source")
        if isinstance(source, list):
            source = ", ".join(source)

        # 🔹 Manually filter by institution name (Ensuring all fields are strings before calling .lower())
        if INSTITUTION.lower() in str(provider).lower() or INSTITUTION.lower() in str(alternative_provider).lower() or INSTITUTION.lower() in str(source).lower():
            results.append({
                "Title": title,
                "Description": description,
                "Video URL": video_link,
                "Provider": provider,
                "Alternative Provider": alternative_provider,
                "Source": source
            })

    # Convert to DataFrame
    df_results = pd.DataFrame(results)

    # Display the DataFrame
    # print(df_results.head())  # Print first few rows

    # Save to CSV
    df_results.to_csv("video_search_results_filtered.csv", index=False)

    print(f"✅ Saved {len(df_results)} results to 'video_search_results_filtered.csv'")

else:
    print("❌ Error:", response.status_code, response.text)
```
